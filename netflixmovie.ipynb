Categorize Vote_Average column

```python
print(df.dtypes)
print(df['Vote_Average'].apply(type).value_counts())
```

```python
def categorize_col(df, col, labels):
    df = df.copy()  # Prevent SettingWithCopyWarning

    # Define quantile values for bin edges
    q1 = df[col].quantile(0.25)
    q2 = df[col].quantile(0.50)
    q3 = df[col].quantile(0.75)
    min_val = df[col].min()
    max_val = df[col].max()

    # Ensure all edges are unique (avoid duplicates in bins)
    edges = sorted(set([min_val, q1, q2, q3, max_val + 0.1]))

    # Apply categorization
    df[col + '_category'] = pd.cut(df[col], bins=edges, labels=labels, duplicates='drop')

    return df

# Labels
labels = ['not_popular', 'below_avg', 'average', 'popular']

# Apply function
df = categorize_col(df, 'Vote_Average', labels)

# Check unique categories
print(df['Vote_Average_category'].unique())
```
